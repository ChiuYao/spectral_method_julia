# Example 1.1 基本函數
這是一個畫出 $\sin(x)$ 的例子。在 plot 指令中我們指定三件事:
1. 告訴他我們要畫 $y = \sin(x)$ 這個函數。
2. 再來要指定函數的範圍 $-2\pi < x < \pi$。
3. 最後用 label 標註函數圖形。注意要把標註的文字放在 " " 中。

另外，也可以畫 $e^x$、$arcsin(x)$ 這些基本函數。

```julia
plot(sin, -pi, pi, label="sine function")
```

# Example 2
接著我們看三維圖形的例子。這個例子中我們自訂 x, y, z 的取法。
1. d = 8*pi/N 的意思是將 $8\pi$ 切成 N 等分
2. 定義 n，n 是一個 201*1 的向量，n = [0, 1, 2, ...,200]。
3. x 值用 $dn \times \cos(\theta)$ 來定義，d 是一個固定的數字。因此 $\cos(\theta)$ 前的係數 dn 會介於 0 到 $8\pi$, $\theta$ 也會介於 0 到 $8\pi$。
4. y 的值與 x 的值定義方法類似。
5. z = [0, 1, 2, ..., 200]，這行是在定義z的範圍

由 (x,y,z) 的定義方式我們知道它是一個螺旋線，螺旋的半徑從 0 一直增加到 $8 \pi$。角度也是從 0 到 $8 \pi$，所以總共繞了4圈。
在 plot 指令中我們用 marker 選擇圖形的樣式為圓圈，legend=false 代表不標記圖形。


```julia
N = 200; d = 8*pi/N;
n= 0.0:N;
x = d*n.*cos.(d*n);
y = d*n.*sin.(d*n);
z= 0.0:N;
plot( x, y, z, marker=:circle, legend=false)
```

# Example 2
接著我們看三維圖形的例子。這個例子中我們要畫出三圍中的一條曲線 $(x(t), y(t), z(t)) = (t\cos t, t\sin t, t)$, $t\in[0, 8\pi]$.

1. 先將 $8\pi$ 切成 N 等分, 間距為 $\Delta t = 8\times\pi/N$.
    1. 在 julia 中可使用 unicode 變數名稱, 比如 $\Delta$ 打法是 `\Delta` 接著按 tab 鍵, $\pi$ 則是 `\Pi`.
2. 定義 $n$ 是一個 $(N+1)\times 1$ 的向量，n = [0, 1, 2, ...,N]。
3. 定義 $ti$ 是 $\Delta t\times n$
3. $x = ti.*\cos.(ti)$，注意向量相乘以及函數在向量的取值都需要加個`.`。
4. y 與 z 的值與 x 的值定義方法類似。

由 (x,y,z) 的定義方式我們知道它是一個螺旋線，螺旋的半徑從 0 一直增加到 $8 \pi$。角度也是從 0 到 $8 \pi$，所以總共繞了4圈。
在 plot 指令中我們用 marker 選擇圖形的樣式為圓圈，legend=false 代表不標記圖形。



```julia
N = 200;
Δt = 8*π/N;
n= 0.0:N;
ti = Δt*n
x = ti.*cos.(ti);
y = ti.*sin.(ti);
z = ti;
plot( x, y, z, marker=:circle, legend=false)
```

# Example 3
這個例子中，我們將學到怎麼在一個畫好的圖形上再加上另一個圖形。
我們嘗試一個簡單的例子，
1. 第一個圖形，x 座標是取 1 到 10 之間的整數，共十個整數。y 軸取 0 到 1 之間的亂數，共十個小數。再來，用 plot 指令將第一個圖形畫出。
2. 第二個圖形，使用相同的 x 座標，函數值 z 取亂數。將第二個函數圖形與第一個圖形重疊，我們需要指令 plot!

```julia
x = 1:10; y = rand(10);
plot(x,y)
z = rand(10)
plot!(x,z)
```

# Example 4
在一個圖形中同時畫出兩個函數，除了上述的方法之外我們還可以這麼做： 直接把兩個函數值定義成一個矩陣
1. 定義 x 是介於 1 到 10 之間的整數。
2. 定義函數值 y 是一個 10*2 的矩陣，矩陣的第一行代表第一個函數的值，而矩陣的第二行代表第二個函數的值。y值的取法為 0 到 1 之間的亂數，為小數。

使用 plot 將兩個函數值同時畫出，用 title 設定圖形標題，lw=3 設定線段寬度。

用 label=["Line 1" "Line 2"] 標記函數名稱，須注意順序：
Line 1 會標記 y 的第一個函數值，Line 2 會標記 y 的第二個函數值。

x 軸的名稱有兩種設定方式，
1. 在 plot 指令中加入：xlabel = "My x label"。 (xlabel 後面不用"!")
2. 或是在 plot 指令的下一行用 xlabel!("My x label") 加入。(xlabel! 後面要用"!")

```julia
x = 1:10; y = rand(10,2);
plot(x,y,title="Two Lines",label=["Line 1" "Line 2"],lw=3,xlabel = "My x label")
```

```julia
x = 1:10; y = rand(10,2);
plot(x,y,title="Two Lines",label=["Line 1" "Line 2"],lw=3)
xlabel!("My x label")
```

# Example 5
在一個圖形中畫出三個函數圖形，我們可以試試在原有的圖形上再畫上另一個函數。
1. 先畫出第一個函數圖形，在這裡我們將第一個畫出來的圖形叫做 p。 p = plot(x,y)
2. 將後來想加入的函數用 plot! 指令加進來，所以 plot! 第一個 input 就要放原本的圖形 p，之後再放入新加入函數的 x 與 z 值。


```julia
x = 1:10; y = rand(10,2);
p = plot(x,y)
z = rand(10)
plot!(p,x,z)
```

# Example 6
如果函數取點的方法不同，能不能畫在同一張圖形中呢？
我們看下面的例子：
1. y 這個矩陣裡包含兩筆資料，x 的取點都是從 1 到 10 中的十個整數。所以可以用 "一個" 指令把兩筆資料畫出來。令這張圖叫做 p
2. z 這筆資料為 0 到 1 中的 19 個亂數，xx 的取點也是19個在 [1, 10] 的均勻取點。所以我們用 plot! 指令，先畫出 p, 再把 z 這筆資料畫上去。

```julia
x = 1:10; y = rand(10,2);
p = plot(x,y)

xx = 1:0.5:10
z = rand(19)
plot!(p,xx,z)
```

利用下列的指令是行不通的!

```julia
x = 1:10; y = rand(10,2);
xx = 1:0.5:10
z = rand(19)
plot(x,y,xx,z)
```
