# Example -- Plotting

底下的幾個例子我們將學習怎麼使用繪圖指令。

```julia
using Plots
```

# Example 1 基本函數
這是一個畫出 $\sin(x)$ 的例子。在 plot 指令中我們指定三件事:
1. 告訴他我們要畫 $y = \sin(x)$ 這個函數。
2. 再來要指定函數的範圍 $-2\pi < x < \pi$。
3. 最後用 label 標註函數圖形。注意要把標註的文字放在 " " 中。

```julia
plot(sin, -1, 1, label="sine function")
```

# Example 2
接著我們看三維圖形的例子。這個例子中我們自訂 x, y, z 的取法。
1. d = 8*pi/N 的意思是將 $8\pi$ 切成 N 等分
2. x 值用 $dn \times \cos(\theta)$ 來定義, d 是一個固定的數字，n 是一個 201*1 的向量, n = [0, 1, 2, ...,200]。因此 $\cos(\theta)$ 前的係數 dn 會介於 0 到 $8\pi$, $\theta$ 也會介於 0 到 $8\pi$.
3. y 的值與 x 的值定義方法類似。
4. z = [0, 1, 2, ..., 200]，這行是在定義z的範圍

由 (x,y,z) 的定義方式我們知道它是一個螺旋線，螺旋的半徑從 0 一直增加到 $8 \pi$。角度也是從 0 到 $8 \pi$，所以總共繞了4圈。
在 plot 指令中我們用 marker 選擇圖形的樣式為圓圈，legend=false 代表不標記圖形。


```julia
N = 200; d = 8*pi/N;
n= 0.0:N;
xx = d*n.*cos.(d*n);
x= [d*n*cos.(d*n) for n in 0:N]; 
y= [d*n*sin.(d*n) for n in 0:N];
z= 0.0:N;
plot( xx, y, z, marker=:circle, legend=false)
```

# Example 3
這個例子中，我們將學到怎麼在一個畫好的圖形上再加上另一個圖形。
我們嘗試一個簡單的例子，
1. 第一個圖形，x 座標是取 1 到 10 之間的整數，y 軸取 0 到 1 之間的亂數。再來，用 plot 指令將第一個圖形畫出。
2. 第二個圖形，使用相同的 x 座標，函數值 z 取亂數。將第二個函數圖形與第一個圖形重疊，我們需要指令 plot!

```julia
x = 1:10; y = rand(10);
plot(x,y)
z = rand(10)
plot!(x,z)
```

# Example 4
在一個圖形中同時畫出兩個函數，除了上述的方法之外我們還可以這麼做： 直接把兩個函數值定義成一個矩陣
1. 定義 x 是介於 1 到 10 之間的整數
2. 定義函數值 y 是一個 10*2 的矩陣，矩陣的第一行代表第一個函數的值，而矩陣的第二行代表第二個函數的值。

使用 plot 將兩個函數值同時畫出，用 title 設定圖形標題，用 label=["Line 1" "Line 2"] 標記函數名稱，lw=3 設定線段寬度。


```julia
x = 1:10; y = rand(10,2);
plot(x,y,title="Two Lines",label=["Line 1" "Line 2"],lw=3)
xlabel!("My x label")
```

# Example 5
在一個圖形中畫出三個函數圖形，我們可以試試在原有的圖形上再畫上另一個函數。
1. 先畫出第一個函數圖形，在這裡我們將第一個畫出來的圖形叫做 p。 p = plot(x,y)
2. 將後來想加入的函數用 plot! 指令加進來，所以 plot! 第一個 input 就要放原本的圖形 p，之後再放入新加入函數的 x 與 z 值。


```julia
x = 1:10; y = rand(10,2);
p = plot(x,y)
z = rand(10)
plot!(p,x,z)
```

```julia
x = 1:10; y = rand(10,2);
plot(x,y,title="Two Lines",label=["Line 1" "Line 2"],lw=3)
xlabel!("My x label")
```
